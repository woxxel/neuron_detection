## Videos

### Recordings
Calcium imaging recordings are obtained as videos / stacks of images in the shape of $X \times Y \times T$ ($X$, $Y$ begin the number of x- and y-pixels, respectively, $T$ being the number of recorded frames.


### Calcium indicators
Pixel intensities scale with the density of calcium-ions. Calcium ions play a central role in the generation of action potentials in neurons: they are actively pumped inside the neuron body when an action potential is generated, and thus show extremely high concentrations. Calcium indicators bind to the calcium-ions and can be stimulated by light to emit photons of a characteristic wave-length. In 1-photon imaging, a single wave is sufficient for stimulation, in 2-photon imaging, the intersection of two beams is necessary, requiring a good focus, but also resulting in a much better resolution. 
A whole range of indicators exist, differing in their sensitivity towards changes in Calcium and their velocity in both responding to a change in Ca$^{2+}$-density and the response decay thereafter.


### What can be seen?
Calcium imaging rasters the image pixel by pixel, and thus displays an imaging window completely. Structures of neurons, axons, neuropil and blood vessels can be visible, though you will mostly see neurons when they become active.
Measurements have some uncertainty in the $z$-plane, resulting in few $\mu$m of Ca$^{2+}$ density being recorded above and below the focus-plane (a typical neuron diameter is $\approx 10\mu$m wide). This can result in seemingly overlapping neuron bodies, causing some pain in the reliable automated extraction of neuron bodies
